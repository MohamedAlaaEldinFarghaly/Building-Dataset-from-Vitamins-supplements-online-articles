```
from bs4 import BeautifulSoup
import urllib.request
import os

try:
  path1="/content/Vitamins_Articles_Texts"
  os.mkdir(path1)
except OSError as error:
  pass

headers = "Vitamins, Article_Link \n"
# opens file, and writes headers
f = open("Links.csv", "w")
f.write(headers)


path = "https://www.rxlist.com"
first_path = "https://www.rxlist.com/supplements/alpha_a.htm"
a = urllib.request.Request(first_path,headers={'User-Agent': 'Mozilla/5.0'})
r=urllib.request.urlopen(a).read()
soup1 = BeautifulSoup(r,'lxml')

Vitamins_links=[]

ch = soup1.find_all("div", {"id": "A_Z"})
for item in ch[0].select("li"):
  character = item.text
  ch_link = item.findAll('a')[0].get('href')
  print(character, ch_link)

  a = urllib.request.Request(path + ch_link,headers={'User-Agent': 'Mozilla/5.0'})
  r=urllib.request.urlopen(a).read()
  soup = BeautifulSoup(r,'lxml')

  for v in soup.find_all("div", {"class": "AZ_results"}):
    for item in v.select("li"):
      name = item.text
      link = item.findAll('a')[0].get('href')
      Vitamins_links.append((name,link))

      f.write(name + ", " + link + ", " + "\n")

      a = urllib.request.Request(link,headers={'User-Agent': 'Mozilla/5.0'})
      r=urllib.request.urlopen(a).read()
      soup2 = BeautifulSoup(r,'lxml')
      Asec= soup2.find_all("section", {"class": "pageSection"})
      text=""
      for s in Asec:
        text+=s.div.get_text()
      os.chdir("/content/Vitamins_Articles_Texts")
      f1 = open(name+".txt", "w")
      f1.write(text)
      f1.close()
      


f.close()
```
