```
from bleach import clean
def strip_html(src, allowed=['b','div','p','li']):
    return clean(src, tags=allowed, strip=True, strip_comments=True)
```

```
from difflib import SequenceMatcher,get_close_matches

def similar(a, b):
    return SequenceMatcher(None, a, b).ratio()
```

```
def check_similar_key(a):
  prop=0  
  key ="sdfsadfaf"


  article_keys =[
  "What other names is","What is",
  "How does",
  "Are there safety concerns?",
  "Dosing considerations for",
  "Are there any interactions with medications?",
  "EFFECTIVE FOR...",
  "LIKELY EFFECTIVE FOR...",
  "POSSIBLY EFFECTIVE FOR...",
  "POSSIBLY INEFFECTIVE FOR...",
  "LIKELY INEFFECTIVE FOR...",
  "INEFFECTIVE FOR...",
  "INSUFFICIENT EVIDENCE TO RATE EFFECTIVENESS FOR..."
  ]
  for i in article_keys:
    larity = similar(a.lower(),i.lower())
    if(larity>prop):
      prop=larity
      key=i
  return key
  
```

```
import pickle
def write_to_file(name,data):
  file = open(name, 'wb')
  pickle.dump(data, file)
  file.close()

def load_from_file(name):
  file = open(name, 'rb')
  data = pickle.load(file)
  file.close()
  return data
```
